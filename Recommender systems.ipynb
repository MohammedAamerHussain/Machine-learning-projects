## Item-Item collaborative filtering

We have to construct a new item by item matrix and determine the relationships between the artists. Using the cosine similarity , we can calculate the correlation between the artits. cosine similarity of 1 means perfect similarity where as cosine similarity of 0 means no correlation .

```python
# Dropping the user column to construct a item-item matrix
artists_matrix = df.drop('user', 1)
```

```python
artists_matrix.head()
```

```python
# Creating a placeholder matrix for item-item similarity
temp_df = pd.DataFrame(index=artists_matrix.columns,columns=artists_matrix.columns)
```

```python
for i in range(0,len(temp_df.columns)) :
    # Looping through each of the artists columns
    for j in range(0,len(temp_df.columns)) :
# Filling in the NaNs in the temp_df or the placeholder matrix with the item-item similarities using the spatial distance with cosine
      temp_df.iloc[i,j] = 1 - cosine(artists_matrix.iloc[:,i],artists_matrix.iloc[:,j])
```

```python
temp_df
```

```python
# Building a close neighbourhood for closest similar items/artists
item_similarity_matrix = pd.DataFrame(index=temp_df.columns,columns=range(1,12))
```

```python
item_similarity_matrix
```

```python
# Looping through the entire item-similarity matrix and filling in the values with the closest neighbourhood items/artists
for i in range(0,len(temp_df.columns)):
    item_similarity_matrix.iloc[i,:11] = temp_df.iloc[0:,i].sort_values(ascending=False)[:11].index
```

```python
item_similarity_df = item_similarity_matrix.transpose()
item_similarity_df
```

```python
# Dropping the first row as it is redundant with the column names
item_similarity_df = item_similarity_df.drop(item_similarity_df.index[0])
```

```python
item_similarity_df.head(10)
```

# User Based collaborative Filtering

```python
# Building a co-occurrence matrix to find how many times items/artists have appeared in the radio songs matrix together
df = pd.read_csv('radio_songs.csv')
user_based_df = df
user_based_df.set_index('user', inplace = True)
original_df = np.array(user_based_df)
transpose_df = np.array(user_based_df.T)
```

```python
# Co-occurence matrix is the cross product between the original matrix and its transpose
co_occurence_matrix = np.dot(transpose_df,original_df)
np.fill_diagonal(co_occurence_matrix, 0)
co_occur_df = pd.DataFrame(co_occurence_matrix, columns = user_based_df.columns, index = user_based_df.columns)
co_occur_df
```

```python
# Building a user-item recommendation matrix first
user_item_matrix = np.dot(original_df, co_occurence_matrix)
non_zero_loc = pd.np.nonzero(original_df)
user_item_matrix[non_zero_loc] = 0
user_recommend_df = pd.DataFrame(user_item_matrix, columns = user_based_df.columns, index = user_based_df.index)
user_recommend_df
```

```python
# Replacing null values in the user recommendation sparse matrix by songs/artists/item Average
artists_final = user_recommend_df.fillna(user_recommend_df.mean(axis=0))  # row-wise average

# Replacing the null values in the user recommendation sparse matrix by user rating Average
user_final = user_recommend_df.apply(lambda row: row.fillna(row.mean()), axis=1)  # column-wise average
```

```python
artists_final
```

```python
user_final
```

```python
# we still see a lot of null values in the sparse matrix
# Building a user similarity matrix using cosine similarity on the user_final sparse matrix
cosine_sim = cosine_similarity(user_final)
np.fill_diagonal(cosine_sim, 0 )
user_cosine_similarity = pd.DataFrame(cosine_sim,index=user_final.index)
user_cosine_similarity.columns=user_final.index
user_cosine_similarity
```

```python
# Building a user similarity matrix using cosine similarity on the artists_final sparse matrix
cos = cosine_similarity(artists_final)
np.fill_diagonal(cos, 0 )
artists_cosine_similarity = pd.DataFrame(cos,index=artists_final.index)
artists_cosine_similarity.columns=user_final.index
artists_cosine_similarity
```

# 2. Conceptual questions:

1. Name 2 other similarity measures that you can use instead of cosine similarity above.

Cosine similarity is one of the most common similarity measure used for the collaborative filtering . Besides cosine similarity , we can also use pearson correlation or Euclidean distance measure as well.

Cosine similarity measure can be applied when we have a sparse matrix(ratings are undefined or null) like the one we just worked on(radio_songs.csv). Cosine similarity normalizes the scales of the features by default because it doesnt care about the magnitude. Its better not to use cosine similarity on ordinal categorical data that is not centered or even binary/unary data which happens in the implicit feedback use cases.

Euclidean distance : We all know that the similar items lie in close proximity or vicinity to each other if we plot them in multi-dimensional space. Euclidean distance helps in calculating the distance between the items based on that distance and then recommend items or products to the users.
Euclidean proximity/distance similarity measure can be used if the input data is not sparse and the values within the data matrix or the values of the features/atrributes are significant. Euclidean measure helps to balance the sensitivity with respect to the vector direction withoit being overtly biased to popular items. Optimizing the recommender systems models using euclidean measure gives more intuitive results.

Pearson correlation : This measure indicates how much two products or items are correlated .Their similarity increases with their similarity
Pearson correlation measure can be selected when ratings or the values of different attrbutes/features are on different scale and also when the input data is subject to user-bias. Pearson correlation also normalizes the scales of the user ratings by default like cosine. Pearson correaltion helps to mitigate the bias towards popular itemsbut also has a drawback of recommending items that are not popular.

We can also select adjusted cosine particularly to adjust for the user-bias specifically for the item-based approach.
Adjusted cosine works by substracting individual user's average rating from each co-rated pair.

2. What is needed to build a Content-Based Recommender system?

Content or the attributes of an item or product is needed to build a content based recommendation system. It works by considering or tagging the products or items using certain keywords in order to understand what the user likes , then query or look up in the database with those keywords and then recommend other different products or items with similar attributes or tags. Content based recommender system helps in predicting the rating or preference that any particular user would give to an item or product.

Content based recommender systems initially need the data that the users provides whether it is explicit like a ratig or implicit like clicking on a link. On the basis of this obtained data, profiles of the users is created, which in turn helps in making suggestions to the user. This recommender system becomes more and more accurate in terms of predictions/recommendations as the users provide more inputs or takes actions on the suggestions/recommendations.

There are 3 basic structures in a content based recommender systems : User profile , item profile and a utility matrix

User profile :  Within user profile , vectors are created based on user's preference 

item profile : Item profile holds the important attributes/tags or charecteristics of the items .

utility matrix : Utility matrix describes the reltionship between the user and the item which is then fed into the content based recommendation system. Utility matrix represents the data obtained from the user and then find the relations between the songs/items/products in terms of which were disliked and liked. A degree of preference is assigned to each user-item pair. This helps in forming a user matrix with respective items in order to identify their preference relationship.

3. Name 2 methods to evaluate your recommender system.


Since we worked on a binary unary matrix , it would make sense if our collaborative filtering recommendation system models are evaluated on precision , recall , f1 score etc.

Precision : Precision would indicate out of all the songs or albums that were recommended , how many of those did user actually like/purchase/listened to ? Precision is given by true positive/(true positive + false positive)
The denominator represents the total number of songs that were recommended to a particular user and the numerator indicates out of all of those recommendations(in the denominator) made , how many did the user actually ended up liking/purchasing/listening .If we have a larger precision , then it means that the recommendations were better or accurate.
Precision will be 0.8 if 5 songs were recommended by the model and out of which the user liked/purchased/listened to 4 of those.


Recall : Recall is a measure of what proportion of songs that a user liked/listened/purchased were actually recommended. Recall is given by : True positive/( true positive + false negative). The true positive part or the numerator indicates the number of songs recommended to a user that she/he likes and then the denominatir represents total number of songs that he/she listened to/purchased/liked. If the recall is bigger , it means our recommendation engine is performing better. 
The recal will have a value of 0.6 if the user likes/puchases/listend to 5 items and out of those 5 , 3 were recommended by the model.


MAP at k (Mean Average Precision at cutoff k): Since recall and precision do not care about the ordering of our song recommendations , precision at cutoff k is the value of the precision that is calculated by taking into consideration only the subset of the recommendations from rank 1 to k.

RMSE and MAE can be applied to non-binary matrices which may not be applicable to our use case but worth mentioning because they are widely used.

RMSE(root mean squared error is the most widely used evaluation metric for the recommender systems. RMSE puts higher penalty or rather punishes more larger errors.

MAE( Mean absolute error) is also another popular metric for meansuring how close is our prediction (of the recommender system) numerically from the real value. MAE punishes or puts penalty on every error regardless of whether its small or big.

```python
# Building a temporary place holder matrix for the user_item similarities
user_item_df = pd.DataFrame(index=df.index,columns=df.columns)
```

```python
user_item_df.iloc[:,:1] = df.iloc[:,:1]
```

```python
user_item_df
```

```python
# Getting similarity score
def get_similarity_score(purchase_history, similarity_scores):
   return sum(purchase_history*similarity_scores)/sum(similarity_scores)
```

```python
# Creating a placeholder matrix for item-item similarity
placeholder_df = pd.DataFrame(index=artists_matrix.columns,columns=artists_matrix.columns)
```

```python
for i in range(0,len(placeholder_df.columns)) :
    # Looping through each of the artists/songs columns
    for j in range(0,len(placeholder_df.columns)) :
# Filling in the NaNs in the temp_df or the placeholder matrix with the item-item similarities using the spatial distance with cosine
      placeholder_df.iloc[i,j] = 1 - cosine(artists_matrix.iloc[:,i],artists_matrix.iloc[:,j])
```

```python
# Building a close neighbourhood for closest similar items/artists
songs_neighbourhood = pd.DataFrame(index=placeholder_df.columns,columns=range(1,11))
```

```python
# Looping through the entire songs_neighbourhood matrix/item-similarity matrix and filling in the values with the closest neighbourhood songs/items/artists
for i in range(0,len(placeholder_df.columns)):
    songs_neighbourhood.iloc[i,:10] = placeholder_df.iloc[0:,i].sort_values(ascending=False)[:10].index
```

```python
songs_neighbourhood
```

```python
# Dropping the first row as it is redundant with the column names
# songs_neighbourhood = songs_neighbourhood.drop(songs_neighbourhood.index[0])
# songs_neighbourhood
```

```python
# Our approach is to find top 10 neighbours of the current song in iteration and then find top 10 closest neighbours using 
#item-item collaborative filtering  then getting the purchase record of the current user for each closest neighbour of the current song

# Looping through all the rows of the placeholder matrix for the user-item similarities and filling in values
for i in range(0,len(user_item_df.index)):
    # For each column or song/artist/item , we have to build a neighbourhood of closest similar songs( top 10)
    for j in range(1,len(user_item_df.columns)):
        user_iter = user_item_df.index[i]
        song_iter = user_item_df.columns[j]
 
        if df.loc[i][j] == 1:
            user_item_df.loc[i][j] = 0
        else:
            current_song_top_neighbours = songs_neighbourhood.loc[song_iter][1:10]
            top_closest_songs = placeholder_df.loc[song_iter].sort_values(ascending=False)[1:10]
            user_purchase_record = artists_matrix.loc[user_iter,current_song_top_neighbours]
 
            user_item_df.loc[i][j] = get_similarity_score(user_purchase_record,top_closest_songs)
```
